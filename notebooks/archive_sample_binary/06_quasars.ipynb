```python
import pandas as pd

for name in ["quasars_gaia", "stars_gaia"]:
    path = f"../data/full/{name}.csv"
    df = pd.read_csv(path)
    print(f"=== {name} ({len(df)} rows, {df.shape[1]} cols) ===")
    miss = df.isna().mean().sort_values(ascending=False)
    print("Top missing (%):")
    print((miss * 100).head(10).round(2))
    print()
```
