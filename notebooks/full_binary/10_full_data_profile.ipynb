```python
import pandas as pd
from pathlib import Path

files = [
    ("blackholes_full", "../data/full/blackholes_gaia.csv"),
    ("stars_full", "../data/full/stars_gaia.csv"),
    ("exoplanets_full", "../data/full/exoplanets.csv"),
    ("comets_full", "../data/full/comets.csv"),
    ("asteroids_full", "../data/full/asteroids.csv"),
]

for name, path in files:
    p = Path(path)
    if not p.exists():
        print(f"{name}: missing {path}")
        continue
    try:
        df = pd.read_csv(p)
    except Exception:
        df = pd.read_csv(p, engine="python", on_bad_lines="skip")
    miss = (df.isna().mean() * 100).sort_values(ascending=False)
    print(f"=== {name} ({len(df)} rows, {df.shape[1]} cols) ===")
    print(miss.head(10).round(2), "\n")
```
